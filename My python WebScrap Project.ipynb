```python
from bs4 import BeautifulSoup
import requests
import smtplib
import datetime
import time
from selenium import webdriver
import csv

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

url = 'https://www.snapdeal.com/search?keyword=tshirt%20men&santizedKeyword=&catId=&categoryId=0&suggested=true&vertical=p&noOfResults=20&searchState=&clickSrc=suggested&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=ALL&url=&utmContent=&dealDetail=&sort=rlvncy'

driver.get(url)

# Scroll down to load more items
scroll_pause_time = 2
while True:
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait for some time to load more items
    time.sleep(scroll_pause_time)
    # Break the loop if no more items are loaded
    if len(BeautifulSoup(driver.page_source, 'html.parser').find_all(class_='product-tuple-listing')) >= 500:
        break

# Parse the loaded page content
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all product containers
product_containers = soup.find_all(class_='product-tuple-listing')

if product_containers:
    with open('Test webscrap datas.csv', 'w', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(['Title', 'Price', 'Rating', 'Number of Ratings', 'Size'])

        for container in product_containers:
            # Extract title
            title_element = container.find(class_='product-title')
            title = title_element.get_text().strip() if title_element else 'N/A'

            # Extract price
            price_element = container.find(class_='lfloat product-price')
            price = price_element.get_text().strip() if price_element else 'N/A'

            # Extract rating
            rating_element = container.find(class_='clearfix rating av-rating')
            rating = 'N/A'
            if rating_element:
                filled_stars = rating_element.find_all(class_='filled-stars')
                if filled_stars:
                    rating_style = filled_stars[0].get('style', '')
                    rating_percentage = float(rating_style.split(':')[1].replace('%', '').strip())
                    rating = rating_percentage / 20  # Convert percentage to a 5-star scale
            # Extract number of ratings
            rating_numbers = container.find(class_='product-rating-count')
            number_ratings = rating_numbers.get_text(strip = True)

            # Extract Size
               
            size_element = container.find(class_='attr-sec marT5')
            size = 'N/A'
            if size_element:
                subattr_values = size_element.find_all(class_='sub-attr-value')
                if subattr_values:
                    size = ', '.join([subattr_value.get_text(strip=True) for subattr_value in subattr_values])

            

            writer.writerow([title, price, rating, number_ratings, size])
else:
    print('Product containers not found')

# Close the WebDriver
driver.quit()
```
